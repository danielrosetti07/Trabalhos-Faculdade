# **O que é o algoritmo Apriori?**




O algoritmo Apriori é um algoritmo de aprendizado de máquina não supervisionado
utilizado no aprendizado de regras de associação. Aprendizado de regras de
associação é uma técnica de mineração de dados que identifica padrões, conexões e dependências frequentes entre diversos grupos de itens chamados conjuntos de itens nos dados. Alguns casos de uso comuns são sistemas de previsão e recomendação, como análise de cestas de mercado para plataformas de comércio eletrônico.

Foi criado por  Rakesh Agrawal e Ramakrishnan Srikant em 1994, é utilizado para encontrar conjuntos de itens frequente que aparecem em um conjunto de dados.
Ele funciona analisando combinações de itens passo a passo (1 item, depois 2, depois 3...), usando a ideia de que:               
* Se um grupo de itens é frequente, todos os seus subconjuntos também são.  
* Se um grupo não é frequente, nenhum conjunto maior com ele será.

É muito usado em análise de compras (ex: recomendar meias para quem compra sapatos).

☑ Vantagens:
* Fácil de entender
* Fácil de aplicar em diferentes áreas

⚠ Desvantagens:
* Pode ser lento e consumir muita memória com muitos dados  

Por isso é recomendado utilizar com outras técnicas

A função Apriori é integrada a muitas linguagens de programação populares, inclusive Python, Java e R, facilitando a integração de regras de associação de qualidade junto com conjuntos de itens frequentes em aplicativos ou sistemas existentes.

# **Como  funciona o algoritmo Apriori**

O algoritmo Apriori funciona em etapas para encontrar grupos de itens frequentes e criar regras de associação:

1. Geração de conjuntos de itens frequentes:  
Identifica itens individuais e combinações que aparecem com frequência eliminando os que não aparecem com frequência.
2. Expansão e posterior poda de conjuntos de itens:  
Combina itens frequentes para formar grupos maiores e remove combinações pouco frequentes, reduzindo o trabalho.
3. Repetição da etapa 1 e 2:  
Repete esse processo até encontrar todos os conjuntos de itens relevantes.

No final, o algoritmo analisa esses conjuntos para descobrir relações fortes entre os itens.

# **Medição dos conjuntos de itens**



#   Support:

    É definido como a frequência que um item aparece no conjunto de itens.
    
    Support {X} = número de vezes que X aparece / quantidade total de itens.

    Essa métrica, portanto, define a probabilidade de ocorrência de cada item individual nas transações.
    A mesma lógica pode ser estendida aos conjuntos de itens.

#    Confidence:
    
    Identifica a probabilidade de itens ou conjuntos de itens ocorrerem juntos nos conjuntos de itens.

    Confidence {X -> Y} = número de vezes que X e Y aparecem / número de vezes que X aparece.

    Ela representa a probabilidade de que os itens em Y sejam comprados quando os itens em X são comprados.

    
#    Lift:

    É o fator que indica o quanto a probabilidade do item *X* levar ao item *Y* é maior do que a probabilidade isolada de *X* ocorrer.

    Lift {X -> Y} = Support {X, Y} / Support {X} * Support {Y}

    Lift > 1: associação positiva (itens comprados juntos com mais frequência)
    Lift = 1: coincidência
    Lift < 1: associação negativa (itens comprados juntos com menos frequência)



# **Aplicando o algoritmo**

```python
import warnings
warnings.filterwarnings('ignore')

# Importando as bibliotecas pandas para manipulação de dados e as funções 'apriori' e 'association_rules'
# do mlxtend para o algoritmo.

import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules

# Definindo um dicionário chamado data, que representa cinco transações de compras, onde cada chave é um produto
# e os valores (1 e 0) indicam se o produto foi comprado ou não.

data = {
    'Leite': [1, 1, 1, 0, 1, 1, 0],
    'Pão': [1, 0, 1, 1, 0, 1, 1],
    'Manteiga': [0, 1, 0, 1, 0, 1, 0],
    'Ovo': [1, 0, 0, 1, 1, 0, 1],
    'Cerveja': [0, 1, 1, 1, 1, 1, 1]
}

# Convertendo o dicionário em um DataFrame do Pandas chamado df, formato ideal para as próximas etapas

df = pd.DataFrame(data)
```

A imagem mostra um processo de detecção de fraude usando regras de associação.
* O sistema analisa dados e descobre padrões suspeitos automaticamente.

1. Dados de treinamento
    * Começa com um conjunto de dados (Training Data Set)
    * Exemplo: registros de compras, transações, eventos

2. Filtragem
    * Os dados são filtrados (Filtering Event Log)
    * Remove o que não é relevante

3. Análises principais
* O sistema analisa os dados de várias formas:

    * Fluxo de atividades (Control Flow)
    * Uso de recursos (Resource Analysis)
    * Tempo de execução (Throughput Time)
    * Pontos de decisão (Decision Point)

4. Identificação de comportamento suspeito

5. Aplicação do Apriori

* Aqui entra o principal:

    * O sistema usa o método Apriori
    * Ele gera regras de associação

  Exemplo: “SE cliente compra X e Y → alta chance de fraude”

6. Verificação por especialista
    * Um humano valida as regras (Expert Verification)

7. Teste e checagem
    * Usa outro conjunto de dados (Testing Data Set)
    * Faz uma verificação de conformidade (Compliance Check)

8. Regras finais

  * As regras são separadas em:

    * Fraud Rules → regras de fraude
    * Legal Rules → regras normais

9. Resultado final
    * O sistema decide se algo é fraude (Fraudulent Status)
    * E pode adicionar novas regras ao sistema

Onde entra o Apriori?
* Ele é usado para:

    * Descobrir padrões frequentes
    * Criar regras do tipo “SE… ENTÃO…”
    * Ajudar a identificar comportamentos suspeitos automaticamente
